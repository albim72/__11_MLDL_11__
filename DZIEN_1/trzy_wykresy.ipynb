```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#tworzenie zakresu wartoÅ›ci -> dziedzina funkcji
x = np.linspace(0,10,100)
```

```python
#definiowanie 3 funkcji
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(-0.1*x)
```

```python
#tworzenie wykresu
plt.plot(x,y1,label='sin(x)',color='blue')
plt.plot(x,y2,label='cos(x)',color='green')
plt.plot(x,y3,label='exp(-0.1*x)',color='red')

plt.title('wykresy trzech funkcji')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)

plt.legend()

plt.show()
```
